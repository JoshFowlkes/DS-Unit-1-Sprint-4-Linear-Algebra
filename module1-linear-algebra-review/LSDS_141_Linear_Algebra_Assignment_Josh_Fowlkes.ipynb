## 3.1 Find $IG$ (be sure to show your work) 

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```
# Inverses, the easy way

G = [[12,11],
     [7,10]
    ]

G = np.array(G)

print('The Inverse:\n', np.linalg.inv(G))

print('\nThe Identity Matrix:\n', np.matmul(G, np.linalg.inv(G)))
```

```
# Inverses, the hard way

G = [[12,11],
    [7,10]]
a = 12
b = 11
c = 7
d = 10

inverse = [[d, -b], [-c,a]]
new = np.array(inverse)

one_over_det = (1/ ((a*d)-(b*c)))
inv_G = np.multiply(one_over_det, new) 

print('The Inverse:\n', inv_G)

print('\nThe Identity Matrix:\n', np.matmul(G, inv_G))



# awesome sauce!! same result on both
```
